```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("2012 State 7 Federal Fatalities (1).csv")
```

```python
df.head(10)
```

```python
#Splitting the original column to have the State names and Geo-coordinates into seperate columns
df["State"].str.split("\n", expand = True)
```

```python
#Renaming Columns with State and Geo-coordinates
df[["State", "Geo-coordinates"]] = df["State"].str.split("\n", expand = True)
```

```python
#Changing the original index of 0 to State column
df.set_index("State")
```

```python
#Total for each type of Program
df["State or Federal Program"].value_counts()
```

```python
#Setting grouping the data frame by program as first_group
first_group = df.groupby("State or Federal Program")
```

```python
#Sum of State
first_group.get_group("State").sum()
```

```python
#Average Years to Inspect
df["Years to Inspect Each Workplace Once"]
df["Years to Inspect Each Workplace Once"].mean()
```

```python
def average_checker(row_value):
    if row_value>df["Years to Inspect Each Workplace Once"].mean():
        return "This value is greater than the average"
    elif row_value==df["Years to Inspect Each Workplace Once"].mean():
        return "This value is equal to the average"
    else:
        return "This value is less than average"
```

```python
#Applying average checker to current column to create a new column
df["my_new_column"]= df["Years to Inspect Each Workplace Once"].apply(average_checker)
```

```python
#Shows whether the value is less than or greater than for Yrs To Inspect Each Workplace Once
df["my_new_column"]
```

```python
#Greater concentration of less years inspected in the workplace 
df["my_new_column"].value_counts().plot(kind="bar", legend= True)
```
